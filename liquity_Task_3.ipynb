```python
from pdf2image import convert_from_path
import pytesseract

# Path to your PDF file
pdf_path = '/content/Jamin Learning Doc - 23-01-2024-Final Draft.docx.pdf'

# Convert PDF to images (one per page)
images = convert_from_path(pdf_path)

# Initialize an empty string to hold the extracted text
extracted_text = ""

# Loop through each image and extract text using pytesseract
for image in images:
    extracted_text += pytesseract.image_to_string(image, lang='guj')  # 'guj' is for Gujarati language

# Output the extracted text
print(extracted_text[:1000])  # Print the first 1000 characters of the extracted text to check
```

```python
# Clean the extracted text to remove unnecessary line breaks and spaces
cleaned_text = extracted_text.replace("\n", " ").strip()  # Replace line breaks with spaces and remove leading/trailing spaces

# Print cleaned text
print(cleaned_text[:1000])  # Print the first 1000 characters to verify
```

```python
import json

# Example function to parse questions and answers
def extract_faq(text):
    faq_list = []

    # Split the text into lines
    lines = text.split(". ")

    for i in range(0, len(lines) - 1, 2):  # Assuming Q and A are in consecutive pairs
        question = lines[i]
        answer = lines[i + 1]
        faq_list.append({"question": question, "answer": answer})

    return {"faq": faq_list}

# Extract FAQ from the cleaned text
faq_data = extract_faq(cleaned_text)

# Save the FAQ data into a JSON file
with open('faq_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(faq_data, json_file, ensure_ascii=False, indent=4)

# Print the JSON data to verify
print(json.dumps(faq_data, ensure_ascii=False, indent=4))
```

```python
import json

# Load the cleaned FAQ data (from the 'corrected_faq_data.json' file)
with open('/content/faq_data.json', 'r', encoding='utf-8') as json_file:
    faq_data = json.load(json_file)

# Function to get an answer from the FAQ data based on the user's query
def get_answer(query):
    for item in faq_data['faq']:
        if item['question'].lower() in query.lower():
            return item['answer']
    return "Sorry, I don't have an answer to that question."

# Simple chatbot loop
while True:
    user_query = input("Ask a question (or type 'exit' to quit): ")
    if user_query.lower() == 'exit':
        break
    response = get_answer(user_query)
    print(f"Answer: {response}")

```
