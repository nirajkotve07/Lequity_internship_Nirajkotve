```python
import os
from PyPDF2 import PdfReader
from pdf2image import convert_from_path
import pytesseract

# Ensure Tesseract is installed and its path is set
pytesseract.pytesseract.tesseract_cmd = r"/usr/bin/tesseract"  # Update based on your system

# Directories

pdf_dir = "pdfs"
text_dir = "extracted_texts"
os.makedirs(text_dir, exist_ok=True)

def extract_text_with_pypdf2(pdf_path):
    """Extract text from a PDF using PyPDF2."""
    try:
        reader = PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            if page:
                text += page.extract_text() or ""
        return text.strip()
    except Exception as e:
        print(f"Error using PyPDF2 for {pdf_path}: {e}")
        return ""

def extract_text_with_ocr(pdf_path):
    """Extract text from a PDF using OCR."""
    try:
        images = convert_from_path(pdf_path)
        text = ""
        for i, image in enumerate(images):
            text += pytesseract.image_to_string(image)
        return text.strip()
    except Exception as e:
        print(f"Error using OCR for {pdf_path}: {e}")
        return ""

# Process all files in the PDF directory
for root, _, files in os.walk(pdf_dir):
    for file in files:
        if file.lower().endswith(".pdf"):
            pdf_path = os.path.join(root, file)
            print(f"Processing: {pdf_path}")

            # Attempt text extraction with PyPDF2 first
            extracted_text = extract_text_with_pypdf2(pdf_path)

            # If PyPDF2 fails to extract meaningful text, fall back to OCR
            if not extracted_text:
                print(f"Falling back to OCR for {file}")
                extracted_text = extract_text_with_ocr(pdf_path)

            # Save the extracted text to a .txt file
            output_file = os.path.join(
                text_dir, f"{os.path.splitext(file)[0]}.txt"
            )
            with open(output_file, "w", encoding="utf-8") as f:
                f.write(extracted_text)

            print(f"Text extraction complete for {file}")

print("All PDFs processed. Extracted texts are saved in the 'extracted_texts' folder.")
```

```python
import os
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Initialize the sentence transformer model
model_name = 'sentence-transformers/all-MiniLM-L6-v2'  # Lightweight model for embeddings
model = SentenceTransformer(model_name)

def split_text_into_chunks(text, max_length=512):
    """Split the extracted text into manageable chunks."""
    words = text.split(' ')
    chunks = []
    chunk = ''

    for word in words:
        if len(chunk + ' ' + word) > max_length:
            chunks.append(chunk.strip())
            chunk = word
        else:
            chunk += ' ' + word
    if chunk:
        chunks.append(chunk.strip())

    return chunks

def generate_embeddings(text):
    """Generate embeddings for the given text."""
    # Split the text into smaller chunks
    text_chunks = split_text_into_chunks(text)

    # Encode each chunk
    chunk_embeddings = model.encode(text_chunks, convert_to_tensor=True)

    # Convert to NumPy and calculate mean embedding
    return np.mean(chunk_embeddings.cpu().numpy(), axis=0)  # Ensure a 1D NumPy array

def load_text_files(text_dir):
    """Load all text files from a directory into a dictionary."""
    texts = {}
    for root, _, files in os.walk(text_dir):
        for file in files:
            if file.lower().endswith(".txt"):
                text_path = os.path.join(root, file)
                with open(text_path, "r", encoding="utf-8") as f:
                    texts[file] = f.read()
    return texts

def generate_document_embeddings(text_dir):
    """Generate embeddings for all documents in a directory."""
    texts = load_text_files(text_dir)
    embeddings = {}
    for file, text in texts.items():
        embeddings[file] = generate_embeddings(text)
    return embeddings

def find_most_relevant_documents(query, embeddings_dict, top_n=3):
    """Find the top N most similar documents to the query."""
    # Generate embedding for the query
    query_embedding = generate_embeddings(query).reshape(1, -1)  # 2D array for cosine similarity

    # Compute similarities
    similarities = {}
    for file, embedding in embeddings_dict.items():
        doc_embedding = np.array(embedding).reshape(1, -1)  # Ensure 2D shape
        similarity = cosine_similarity(query_embedding, doc_embedding)[0][0]
        similarities[file] = similarity

    # Sort by similarity score in descending order
    sorted_docs = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    return sorted_docs[:top_n]

# Paths and setup
text_dir = "extracted_texts"  # Directory containing text files

# Generate document embeddings
print("Generating document embeddings...")
embeddings_dict = generate_document_embeddings(text_dir)
print("Document embeddings generated.")

# Save embeddings to a file
embeddings_file = "embeddings.json"
with open(embeddings_file, "w", encoding="utf-8") as f:
    json.dump({k: v.tolist() for k, v in embeddings_dict.items()}, f, ensure_ascii=False, indent=4)
print(f"Embeddings saved to {embeddings_file}.")

# Query the system
query = "What is the legal principle related to breach of contract?"
top_documents = find_most_relevant_documents(query, embeddings_dict)
print("Top relevant documents:", top_documents)
```

```python
import json
import os
from sentence_transformers import SentenceTransformer

# Path to the directory containing the extracted text files
extracted_text_dir = '/content/extracted_texts'
embeddings_file = '/content/embeddings.json'

# Load the documents from the extracted text files
def load_documents_from_directory(directory):
    documents = []
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        if file_path.endswith('.txt'):
            with open(file_path, 'r') as file:
                content = file.read()
            documents.append((filename, content))
    return documents

# Load the embeddings from the JSON file
def load_embeddings(embeddings_file):
    with open(embeddings_file, 'r') as file:
        embeddings_data = json.load(file)
    return embeddings_data

# Load documents and embeddings
documents = load_documents_from_directory(extracted_text_dir)
embeddings_data = load_embeddings(embeddings_file)

# Check if the number of embeddings matches the number of documents
if len(documents) != len(embeddings_data):
    print("Mismatch between the number of documents and embeddings.")
else:
    print(f"Loaded {len(documents)} documents and {len(embeddings_data)} embeddings.")
```

```python

import faiss
import numpy as np

# Convert the embeddings data into a numpy array
embedding_matrix = np.array([embedding for embedding in embeddings_data.values()])

# Initialize the FAISS index
dimension = embedding_matrix.shape[1]  # Dimensionality of the embeddings
index = faiss.IndexFlatL2(dimension)  # L2 (Euclidean) distance index

# Add embeddings to the FAISS index
index.add(embedding_matrix)

print("Vector store (FAISS index) created successfully.")

# Now you can perform similarity search with the vector store
def search_query(query, top_k=5):
    # Convert query to embedding using the same model
    query_embedding = model.encode([query])

    # Perform the search in the FAISS index
    D, I = index.search(query_embedding, top_k)

    # Get the results
    results = []
    for i in range(top_k):
        doc_id = list(embeddings_data.keys())[I[0][i]]
        results.append((doc_id, D[0][i]))
    return results

# Test the search functionality
query ="What is the legal principle related to breach of contract?"
results = search_query(query)

print("Search results:")
for doc_id, score in results:
    print(f"Document: {doc_id}, Score: {score}")
```
